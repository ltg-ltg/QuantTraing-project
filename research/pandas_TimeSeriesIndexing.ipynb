```python
import pandas as pd
import yfinance as yf
```

```python
# collecting data and checking index type
df = yf.download('AAPL',start='2024-01-01',end='2026-01-01')
print(type(df.index))
print(df.index.dtype)
print(df.index[:3])
print(df.loc['2024-01-02'])
print(df.columns)
```

#### DatetimeIndex

It is a special index class that uses the Timestamp object array as a row label of the Pandas dataframe. It is stored internally in the type of datatime64[ns], so memory efficiency is high, and time computation is possible based on this.

```python
# Comparison Partial String Indexing
print(len(df.loc['2024']))
print(len(df.loc['2024-06']))
print(len(df.loc['2024-01-01':'2024-03-31']))
print(df.loc['2024-01-01':'2024-01-04'])
print(df.iloc[:3])
```

#### "Precision" Principles of Partial String Indexing

- The end value: In Python list slicing or Pandas .iloc() slicing, the end value is not included in a section. '[start:end)'.
On the other hand, in Pandas .loc(), the end value is inclued in a section. '[start:end]'.
With the data aligned, it can be quickly sliced based on binary search by using .loc()
- Expand precision: Even if a user inputs a date only up to the year or month like '2024' or '2024-06', Pandas internally extends it to a specific range of the most precise time points (lower and upper limits in nanoseconds)

```python
# resample()
monthly_close = df['Close']['AAPL'].resample('ME').mean() # Series, closed='right', label='right'
print(monthly_close.head(10))
print(type(monthly_close.index))
```

#### Principle of operation of the resample()

- The DateTimeIndex-based resample() operates internally by dividing the time interval by bin and aggregating values within each bin.
- The default of closed and label of 'ME', which means the end of the month, is 'right'. The closed='right' means to include the right end of the period when resampling. '(start,end]'. The label='right' means to label it as the last day of the period.
On the other hand, D, h, min, s have the default of closed='left' '[start,end)', label='left'.

```python
# rolling(): 
rolling_mean = df['Close']['AAPL'].rolling(window=20).mean() # Series, Rolling statistics-based trend extraction
print(rolling_mean.head(25))
```
