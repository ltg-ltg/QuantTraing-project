```python
import yfinance as yf
from fredapi import Fred
import os
from dotenv import load_dotenv
```

```python
# collecting stock price data
spy = yf.download('SPY', start='2025-01-01', end='2025-12-31') # S&P 500 ETF
print(spy.head())
print(spy.shape)
print(spy.loc['2025-03'])
```

#### Columns of the DataFrame
1) 'Close': It means the closing price (04:00 p.m) of the day.
2) 'High': It means the high price of the day.
3) 'Low': It means the low price of the day.
4) 'Open': It means the opening price (09:30 a.m) of the day.
5) 'Volume': It means the trading volume of the day.

```python
# collecting macroeconomic data
load_dotenv()
fred = Fred(api_key=os.getenv('FRED_API_KEY'))
unrate = fred.get_series('UNRATE', observation_start='2025-01-01')
fedfunds = fred.get_series('FEDFUNDS', observation_start='2025-01-01')
print(unrate.head())
print(unrate.idxmax(), unrate.max())
print(fedfunds.head())
print(fedfunds.idxmax(), fedfunds.max())
```

#### Representative Series ID
1) 'FEDFUNDS': Effective Federal Funds Rate
2) 'DGS10': U.S. 10-year Treasury yield
3) 'T10Y2Y': 10-2-year Treasury bond short and long term interest rate differential
4) 'CPIAUCSL': CPI, Consumer Price Index
5) 'PCEPI': PCE
6) 'UNRATE': Unemployment Rate
7) 'GDPC1': Real Gross Domestic Product
8) 'VIXCLS': CBOE Volatility Index

```python
# auto_adjusted
spy_adj_False = yf.download('SPY', start='2025-01-01', end='2025-12-31', auto_adjust=False) # default: auto_adjust=True
print(spy_adj_False.head())
print(spy.head())
```

#### Adjusted Close

Adjusted Close is a price adjusted to reflect corporate events such as dividends and stock splits.
The default is 'auto_adjust=True', which means that the data has already been adjusted. When user inputs 'auto_adjust=False', the 'Adj Close' column is added to the DataFrame separately from Close, which means the original closing price.
